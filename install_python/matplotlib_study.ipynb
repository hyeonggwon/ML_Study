```python
# red 별모양, dash 선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기
theta = np.linspace(0, 2 * np.pi, 20)
plt.plot(theta, np.sin(theta), "r*--", markersize=10, label="sin")
# plt.figure() # 새창 열기, 이거 없으면 겹쳐 그리기
plt.plot(theta, np.cos(theta), "go-", markersize=10, label="cos")
plt.xlabel(r"$\theta$")  # r은 raw
plt.ylabel("value")
plt.title("sin and cos", color="b")
plt.legend()
plt.grid()
# plt.xlim([0,np.pi])
# plt.xlim([-1.5,1.5])
plt.axis([0, np.pi, -1.5, 1.5])  # x축 범위, y축 범위
```

### 2개 변수 함수, 등고선으로 보기, 3D로 보기

```python
a = np.linspace(-1, 1, 100)
b = np.linspace(-1, 1, 100)
A, B = np.meshgrid(a, b)  # a와 b의 모든 조합을 만들어줌
L = A**2 + B**2
plt.contour(a, b, L, 20)  # 등고선 그리기, 20은 등고선의 개수
plt.xlabel("a")
plt.ylabel("b")
plt.grid()
```

```python
plt.figure(figsize=(10, 9))  # (가로, 세로)
ax = plt.axes(projection="3d")
ax.plot_surface(A, B, L)

plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
ax.plot_wireframe(A, B, L)
```

```python
figure = plt.figure(figsize=(10, 9))  # (가로, 세로)
ax = figure.add_subplot(projection="3d")
ax.view_init(elev=10, azim=-45)
ax.plot_surface(A, B, L)

figure = plt.figure(figsize=(10, 9))
ax = figure.add_subplot(projection="3d")
ax.view_init(elev=10, azim=-45)
ax.plot_wireframe(A, B, L)
```

```python
from matplotlib.colors import LightSource

plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
ls = LightSource(altdeg=10, azdeg=-45)
rgb = ls.shade(L, plt.cm.RdYlBu)
ax.plot_surface(A, B, L, facecolors=rgb)
ax.plot_surface(A, B, L, cmap="cividis")
# 'viridis', 'plasma', 'inferno', 'magma', 'cividis' 등등 다양한 컬러맵이 있음
```

```python
from matplotlib.colors import LightSource

plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
ls = LightSource(altdeg=10, azdeg=-45)
rgb = ls.shade(
    L, cmap=plt.cm.gist_earth, vert_exag=0.1, blend_mode="soft"
)  # vert_exag는 높이의 과장 정도, blend_mode는 색상 혼합 방식
ax.plot_surface(
    A,
    B,
    L,
    facecolors=rgb,
    rstride=1,
    cstride=1,
    linewidth=0,
    antialiased=False,
    shade=False,
)
```

```python
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 10)
A, B = np.meshgrid(a, b)  # a와 b의 모든 조합을 만들어줌
L = A**2 + B**2

plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
ax.plot(
    A.reshape(-1, 1).squeeze(),
    B.reshape(-1, 1).squeeze(),
    L.reshape(-1, 1).squeeze(),
    "ro",
)

plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
# ax.scatter(A,B,L, s=100, c="g", marker="*", edgecolors="r")
ax.scatter(A, B, L, s=100, c="g", marker="*", alpha=0.3)
```

```python
plt.figure(figsize=(10, 9))
ax = plt.axes(projection="3d")
ax.plot_surface(A, B, L, alpha=0.3, cmap="viridis")
ax.scatter(A, B, L, s=100, c="r", marker="*")
```

### plotly

```python
import plotly.graph_objects as go

a = np.linspace(-2, 2, 100)
b = np.linspace(-2, 2, 100)
A, B = np.meshgrid(a, b)
L = np.cos(A**2.4) + np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(
    contours_z=dict(
        show=True, usecolormap=True, highlightcolor="limegreen", project_z=True
    )
)
fig.update_layout(title="cos(A^2.4)+sin(B^2.4)", width=700, height=600)
```

```python
fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, opacity=0.5)])
fig.update_traces(
    contours_z=dict(
        show=True, usecolormap=True, highlightcolor="limegreen", project_z=True
    )
)
fig.update_layout(
    title="cos(A^2.4)+sin(B^2.4)",
    width=700,
    height=600,
    scene=dict(zaxis=dict(nticks=40, range=[-3, 3])),
)
```
